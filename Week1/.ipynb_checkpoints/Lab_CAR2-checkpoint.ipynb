### Read in the csv file

I'm going to call my df mca, short for marketing_customer_analysis

```python
mca = pd.read_csv('marketing_customer_analysis.csv')
```

### Show me the shape

```python
mca.shape

#my df mca has 9134 rows and 24 columns
```

### Standardize header names

```python
#I want to see all the column header to see if we need to standardize them

mca.columns

#they look ok, but for the sake of it, I will make them all lower case 
#and replace the spaces with _
```

```python
col_lower = []
for i in range(len(mca.columns)):
    col_lower.append(mca.columns[i].lower().replace(" ","_"))
col_lower
```

```python
mca.columns = col_lower
```

```python
mca.head()
```

### Datetime format

Extract the months from the dataset and store in a separate column. 

```python
import datetime

# we need datetime for this so we need to import it
```

```python
mca['month'] = pd.DatetimeIndex(mca['effective_to_date']).month
mca.head()

#we create a new column called month
# we use the function pd.datetimeindex on the column 'effective_to_date' and chose month
```

it worked as we can see the new column month and we have now 25 instead of 24 columns

Now we filter the data to show only the information for the first quarter , ie Jan Feb and March

```python
filtered = mca.query('month < 5')

#in order to get only the values of the first quarter, i need to look at all under 4
```

```python
filtered['month'].unique()

#to check if it worked i looked for the unique values in column month
# there are only 1 and 2 left, so it worked
```

```python
mca['month'].unique()

#because i'm curious i want to see, which months are in the non filtered df
#turns out there are anyway only Jan and Feb entries ... need not to have filtered
```
