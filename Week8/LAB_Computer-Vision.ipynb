I have found an interesting [article](https://www.nature.com/articles/d41586-020-03187-3) on the topic of ethics in computer vision applications.

I mean: is it ok to develop something (e.g. an algorythmn) just for the sake of creating something scientifically new. <br/>To move forward with technology just to proof we can without keeping in mind if that is something that could be taken out of the theoretical world of science into the "real" world and be used for immoral thing. Like they say in the article: using facial recognition technology to distinguish between minorities living in China. Scientifically this might be interesting, BUT you have to keep in mind, that people could use this to harm these minorities further.

<br/><br/>I think scientists have a certain responsibility to keep in mind for what purpose their tools/research might be used outside the scientific world before developing said tool.
